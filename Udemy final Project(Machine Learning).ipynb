```python
years= flight['year'].unique() 
flight
```

```python
for year in years:
    yearly_data = flight[flight['year'] == year]
    plt.plot(yearly_data['Date'], yearly_data['passengers'], label=year)
plt.title("Trend of monthly passenger traffic for all airline")
plt.ylabel("Passenger")
plt.xlabel("Date")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(flight['Date'], flight['passengers'], marker='o', linestyle='-', color='green')
plt.title('Overall Monthly Passenger Traffic Trend (Aggregated)')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```
